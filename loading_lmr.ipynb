# Load LMRv2.1

```python
import pandas as pd
import cfr
import xarray as xr
import os
import matplotlib.pyplot as plt
import pyleoclim as pyleo
```

```python
# Load data using Pangeo-Forge
store = "https://ncsa.osn.xsede.org/Pangeo/pangeo-forge/test/pangeo-forge/staged-recipes/recipe-run-1200/LMRv2p1_MCruns_ensemble_gridded.zarr"
lmr_ds_gridded = xr.open_dataset(store, engine='zarr', chunks={})
lmr_ds_gridded
```

```python
lmr_ds_gridded.to_netcdf('./lmr_ds_gridded.nc')
```

```python
grid_point_paris = lmr_ds_gridded.air_mean.isel(lat=49, lon=2)
```

```python
traces = []

# Create a Series object for each ensemble member
for i in range(grid_point_paris.shape[1]):  # Loop through MCrun dimension
    # Extract time values and convert to list/array if needed
    time_values = grid_point_paris.time.values
    # Extract values for this ensemble member
    values = grid_point_paris[:, i].values
    
    # Create a Series object
    series = pyleo.Series(
        time=time_values,
        value=values,
        time_name='Time',
        time_unit='year',  # Adjust if your time unit is different
        value_name='Air Temperature',
        value_unit='°C',  # Adjust if your temperature unit is different
        label=f'Ensemble Member {i+1}'
    )
    traces.append(series)

# Create an EnsembleSeries object
ensemble = pyleo.EnsembleSeries(traces)

# Create visualization
# Plot all traces
fig, ax = ensemble.plot_traces(num_traces=20)  # Show 20 random traces
ax.set_title('Ensemble Members at Grid Point (49,2)-- Paris')
```
